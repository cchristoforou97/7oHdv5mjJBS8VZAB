```python
## Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns

## ignore warnings
import warnings
warnings.filterwarnings("ignore",module = "matplotlib\..*")
```

```python
## Load data
data = pd.read_csv('data.csv')
data.head()
```

```python
data = data.drop('Unnamed: 0',axis=1)
data.columns
```

```python
## split the data frame into the response variable, y, and the design matrix, X
df = data.copy()
y = df['Happy'] ## response 
X = df.drop('Happy',axis=1) ## design matrix
from sklearn.model_selection import train_test_split
# split the data into training and test data sets
X_train, X_test, y_train, y_test = train_test_split(X, ## data-features to split
                                                    y, ## data-response to split
                                                    test_size=0.2, ## we will leave 20% to test our models on later
                                                    random_state=42, ## make our work reproducible
                                                    shuffle=True) ## prevent data ordering affecting our models
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

clf = make_pipeline(StandardScaler(), SVC(kernel='linear'))
scores = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
print(scores)
```

```python
clf.fit(X_train, y_train)
clf.predict(X_test)
clf.score(X_test, y_test)
```
