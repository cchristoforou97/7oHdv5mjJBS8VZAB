```python
## Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns

## ignore warnings
import warnings
warnings.filterwarnings("ignore",module = "matplotlib\..*")
```

```python
## Load data
data = pd.read_csv('data.csv')
data.head()
```

```python
data = data.drop('Unnamed: 0',axis=1)
data.columns
```

```python
data_logistic = data.copy()
y = data_logistic['Happy'] ## response 
X = data_logistic.drop('Happy',axis=1) ## design matrix

from sklearn.model_selection import train_test_split

# split the data into training and test data sets
X_train, X_test, y_train, y_test = train_test_split(X, ## data-features to split
                                                    y, ## data-response to split
                                                    test_size=0.1, ## we will leave 10% to test our models on later
                                                    random_state=42, ## make our work reproducible
                                                    shuffle=True) ## prevent data ordering affecting our models
```

```python
from sklearn.linear_model import LogisticRegressionCV
clf = LogisticRegressionCV(cv=10, random_state=42).fit(X_train, y_train)
clf.predict(X_test)
clf.score(X_test, y_test)
```

```python
print(f'accuracy: {((clf.predict(X_test) == y_test).sum()) / len(y_test)}')
```
